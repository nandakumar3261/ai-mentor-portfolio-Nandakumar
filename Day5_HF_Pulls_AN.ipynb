```python
from transformers import pipeline
import time

# Local model download
classifier = pipeline(
    'zero-shot-classification',
    model='facebook/bart-large-mnli'
)

# Test data
resumes = [
    'Built React dashboards for 3 startups',
    'Implemented Spring Boot microservices in Java for fintech app',
    'Trained CNN for image classification using PyTorch, 87% accuracy',
    'Cleaned 100k row dataset using pandas + plotted in seaborn',
    'Wrote SQL queries against PostgreSQL'
]

labels = [
    'frontend dev',
    'backend dev',
    'data analyst',
    'ML engineer'
]

# Run classification
start = time.time()

for r in resumes:

    result = classifier(
        r,
        candidate_labels=labels
    )

    print(
        f'{r[:50]:50} '
        f'-> {result["labels"][0]} '
        f'({result["scores"][0]:.2f})'
    )

print(
    f'\nLocal inference time: '
    f'{time.time()-start:.2f}s'
)
```
